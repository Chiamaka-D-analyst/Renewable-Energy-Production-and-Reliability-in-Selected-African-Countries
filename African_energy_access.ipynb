```python
import pandas as pd

df = pd.read_csv("african_energy_access.csv")
df.head()
```

```python
df.columns
```

```python
df
```

```python
energy_mix = df.groupby(["Country", "Energy_Source"])["Energy_MWh"].sum()
energy_mix
```

```python
import matplotlib.pyplot as plt

energy_mix.unstack().plot(kind="bar")
plt.ylabel("Energy Production (MWh)")
plt.title("Energy Mix by Country (Africa)")
plt.show()
```
